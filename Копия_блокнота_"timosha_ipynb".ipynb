```python
import math as math 
import matplotlib.pyplot as plt
import random
import pandas as pd
%matplotlib
```

```python
x = list(range(-100,110,1))
print(x)
x = list(map(lambda el: el/100,x))
y = list(map(lambda el: 2*math.sin(math.pi*3*el),x))
print(x)
print(y)
plt.plot(x,y)
plt.show()
```
