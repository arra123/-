```python
import math as math 
import matplotlib.pyplot as plt
import random
import pandas as pd
%matplotlib inline
```

```python
psg = pd.read_csv('./titanic.csv')
psg
```

```python
a_s = psg[(psg['Age'] > 30) & (psg['Age'] < 40) & (psg['Sex'] == 'male')]
a_s
```

```python
a_s = psg[(psg['Pclass'] == 1) & (psg['Survived'] == 1) | (psg['Sex'] == 'famale') & (psg['Age'] < 30)]
a_s
```

```python
class_survival = psg[["Sex", "Pclass"]]
c_s = class_survival.groupby(["Sex", "Pclass"]).size()
c_s
c_s.plot.bar()
```

```python
a_s = psg[(psg['Age'] < 25)]
class_survival = psg[["Sex","Survived", "Pclass"]]
a_s = class_survival.groupby(["Sex","Survived", "Pclass"]).size()
a_s.plot.bar()
```

```python
x = list(range(-100,110,1)
print(x)
x = list(map(lambda el: el/100,x))
y = list(map(lambda el: 2*math.sin(math.pi*3*el),x))
print(x)
print(y)
plt.plot(x,y)
plt.show()
```
